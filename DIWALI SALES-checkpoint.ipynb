```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df[['User_ID', 'Cust_name', 'Product_ID', 'Gender', 'Age Group', 'Age',
       'Marital_Status', 'State', 'Zone', 'Occupation', 'Product_Category',
       'Orders', 'Amount', 
    #'Status', 
    #'unnamed1'
   ]]
```

```python
df.drop(labels = ['Status', 'unnamed1'], axis = 1, inplace = True)
```

```python
df['Gender'].replace({'M': 1, 'F': 0}, inplace = True)
```

```python
df.columns
```

```python
avg = np.mean(df['Amount'])
```

```python
df.dropna(how = 'all', axis = 0, inplace = True)
```

```python
df.isna().sum()
```

```python
df['Amount'].fillna(value = avg, inplace = True)
```

```python
data = sns.barplot(x = 'Zone', y = 'Amount', hue = 'Gender', data = df, palette = 'rocket')
data.set_title('Zone VS Amount')
sns.set(rc={"figure.figsize":(15, 8)})
for bars in data.containers:
    data.bar_label(bars)
```

```python
data = sns.barplot(x = 'Orders', y = 'Amount', hue = 'Gender', data = df, palette = 'rocket')
data.set_title('Orders VS Amount')
sns.set(rc={"figure.figsize":(20, 8)})
for bars in data.containers:
    data.bar_label(bars)
```

```python
ax = sns.countplot(x = 'Orders', hue = 'Gender', data = df, palette = 'rocket')
sns.set(rc={'figure.figsize':(10,12)})
for data in ax.containers:
    ax.bar_label(data)
```

```python
df.shape
```

```python
sales_genAmount =  df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
ax = sns.barplot(x = 'Gender', y = 'Amount', data = sales_genAmount, palette = 'rocket')
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(5,8)})
ax.set_title('Gender vs Amount')
```

```python
sales_amount = df.groupby(['Orders'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
ax = sns.barplot(x = 'Orders', y = 'Amount', data = sales_amount, palette = 'rocket')
sns.set(rc={'figure.figsize':(20,15)})
ax.set_title('Orders vs Amount')
for bars in ax.containers:
    ax.bar_label(bars)
```
